```python
# REVENUE FEATURES
import pandas as pd

train = pd.read_csv('training_set_final.csv')
test= pd.read_csv('testing_set_final.csv')

def mean_rev_lv1(train_data, test_data):
    mean = train_data[["lv1", "revenue"]].groupby(["lv1"]).mean().reset_index()
    mean = mean.rename(columns={'revenue':'mean_revenue_for_level1'})
    train_data = train_data.merge(mean, how="left", on=["lv1"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    train_data = train_data.drop_duplicates(subset=['store_id'], keep='first')

    # For test
    test_data = test_data.merge(mean, how="left", on=["lv1"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    test_data = test_data.drop_duplicates(subset=['store_id'], keep='first')
    return train_data, test_data

def mean_rev_lv2(train_data, test_data):
    mean = train_data[["lv2", "revenue"]].groupby(["lv2"]).mean().reset_index()
    mean = mean.rename(columns={'revenue':'mean_revenue_for_level2'})
    train_data = train_data.merge(mean, how="left", on=["lv2"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    train_data = train_data.drop_duplicates(subset=['store_id'], keep='first')

    # For test
    test_data = test_data.merge(mean, how="left", on=["lv2"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    test_data = test_data.drop_duplicates(subset=['store_id'], keep='first')
    return train_data, test_data

def mean_rev_lv3(train_data, test_data):
    mean = train_data[["lv3", "revenue"]].groupby(["lv3"]).mean().reset_index()
    mean = mean.rename(columns={'revenue':'mean_revenue_for_level3'})
    train_data = train_data.merge(mean, how="left", on=["lv3"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    train_data = train_data.drop_duplicates(subset=['store_id'], keep='first')

    # For test
    test_data = test_data.merge(mean, how="left", on=["lv3"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    test_data = test_data.drop_duplicates(subset=['store_id'], keep='first')
    return train_data, test_data

def mean_rev_chain(train_data, test_data):
    mean = train_data[["chain_name", "revenue"]].groupby(["chain_name"]).mean().reset_index()
    mean = mean.rename(columns={'revenue':'mean_revenue_chain'})
    train_data = train_data.merge(mean, how="left", on=["chain_name"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    train_data = train_data.drop_duplicates(subset=['store_id'], keep='first')

    # For test
    test_data = test_data.merge(mean, how="left", on=["chain_name"])
    # we get a bunch of duplicates of store_ids...? Remove them.
    test_data = test_data.drop_duplicates(subset=['store_id'], keep='first')
    return train_data, test_data
```
